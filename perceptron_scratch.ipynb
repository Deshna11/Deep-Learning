```python
import pandas as pd 
import numpy as np
from sklearn.datasets import make_classification
X, y = make_classification(n_samples= 100, n_features= 2, n_informative= 1, n_classes=2, n_clusters_per_class= 1,n_redundant=0, random_state= 41,  hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, s = 50, cmap='winter')
```

```python
def step(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
import random
# will return coeff and intercept:
def perceptron(X, y):
    lr = 0.1
    X = np.insert(X, 0, 1, axis= 1)
    weights = np.ones(X.shape[1])
    for i in range(1000):
        j = np.random.randint(0, X.shape[0])
        y_cap = step(np.dot(weights, X[j]))
        weights = weights + lr*(y[j]- y_cap)*X[j]
    return weights[0], weights[1:] 
```

```python
intercept, coeff = perceptron(X, y)
print(intercept)
print(coeff)
```

```python
m = -(coeff[0]/coeff[1])
b = -(intercept/coeff[1])
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
